```python
import pandas as pd
data2023_08 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATASET/apartments_pl_2023_08.csv')
data2023_09 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATASET/apartments_pl_2023_09.csv')
data2023_10 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATASET/apartments_pl_2023_10.csv')
```

```python
#merging 3 dataframes to 1
data = [data2023_08, data2023_09, data2023_10]
df = pd.concat(data)
print(df.shape)
df.describe()
df
```

```python
#removing the duplicates due to the fact that flat sale offer might be still avaiable in 2/3 month in a row
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
#creating new column for hypothesis purpose
df['priceForSquareMeter']=df['price']/df['squareMeters']
df
```

```python
df=df.dropna(subset=['hasElevator','hasBalcony','hasParkingSpace']) #droping NaN data in order to clean the data
```

```python
#creating new dataframes with different features status (is feature included or excluded in df)
filtered_df_all = df.drop(df[(df['hasElevator'] == 'yes') | (df['hasParkingSpace'] == 'yes') | (df['hasBalcony'] == 'yes') ].index)
filtered_df_bps = df.drop(df[(df['hasElevator'] == 'no') | (df['hasParkingSpace'] == 'yes') | (df['hasBalcony'] == 'yes')].index)
filtered_df_eps = df.drop(df[(df['hasElevator'] == 'yes') | (df['hasParkingSpace'] == 'yes') | (df['hasBalcony'] == 'no')].index)
filtered_df_eb = df.drop(df[(df['hasElevator'] == 'yes') | (df['hasParkingSpace'] == 'no') | (df['hasBalcony'] == 'yes')].index)
filtered_df_none = df.drop(df[(df['hasElevator'] == 'no') | (df['hasParkingSpace'] == 'no') | (df['hasBalcony'] == 'no')].index)
filtered_df_all
```

```python
#in order to do a plot analysis it is crucial to get some specific value of the column we want to check, therefore we use mean value for the price per square meter
mean_no_all=filtered_df_all['priceForSquareMeter'].mean()
mean_yes_elev=filtered_df_bps['priceForSquareMeter'].mean()
mean_yes_balcony=filtered_df_eps['priceForSquareMeter'].mean()
mean_yes_parkingspc=filtered_df_eb['priceForSquareMeter'].mean()
mean_yes_all=filtered_df_none['priceForSquareMeter'].mean()
mean_list=[mean_no_all,mean_yes_elev,mean_yes_balcony,mean_yes_parkingspc,mean_yes_all]
print(mean_list)
```

```python
x=['Has 0 features','Has only elevator','Has only balcony','Has only parking space','Has all of the features']
```

```python
import matplotlib.pyplot as plt
color = ['#98e06e', '#a8713d', '#6d9ed6', '#ba6aa5','#2dbdb2']
fig, ax = plt.subplots()
ax.bar(x, mean_list,color=color, align='center', width=0.3,edgecolor='#0d0d07ba')

ax.set_axisbelow(True)
ax.set_xticklabels(x,rotation=15)
ax.grid(color='gray', linestyle='dashed',axis='y')



plt.xlabel("Type of the condidtion")
plt.ylabel("Price per square meter")
plt.title("Comparison of the price per square meter with different features included in the flat")
plt.show()
```

```python
#preparing the data to compare the price per square meter for flats below 35m2
import math
none_df_ls=filtered_df_all.drop(filtered_df_all[filtered_df_all['squareMeters'] >= 35].index)
elev_df_ls=filtered_df_bps.drop(filtered_df_bps[filtered_df_bps['squareMeters'] >= 35].index)
balc_df_ls=filtered_df_eps.drop(filtered_df_eps[filtered_df_eps['squareMeters'] >= 35].index)
parkspc_df_ls=filtered_df_eb.drop(filtered_df_eb[filtered_df_eb['squareMeters'] >= 35].index)
all_df_ls=filtered_df_none.drop(filtered_df_none[filtered_df_none['squareMeters'] >= 35].index)
```

```python
#calculating the mean value of price per square meter for analysis
mean_no_all_ls=none_df_ls['priceForSquareMeter'].mean()
mean_yes_elev_ls=elev_df_ls['priceForSquareMeter'].mean()
mean_yes_balcony_ls=balc_df_ls['priceForSquareMeter'].mean()
mean_yes_parkingspc_ls=parkspc_df_ls['priceForSquareMeter'].mean()
mean_yes_all_ls=all_df_ls['priceForSquareMeter'].mean()
mean_list_ls=[mean_no_all_ls,mean_yes_elev_ls,mean_yes_balcony_ls,mean_yes_parkingspc_ls,mean_yes_all_ls]
print(mean_list_ls)
```
