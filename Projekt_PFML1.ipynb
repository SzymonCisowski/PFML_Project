```python
import pandas as pd
data2023_08 = pd.read_csv('drive/MyDrive/apartments_pl_2023_08.csv')
data2023_09 = pd.read_csv('drive/MyDrive/apartments_pl_2023_09.csv')
data2023_10 = pd.read_csv('drive/MyDrive/apartments_pl_2023_10.csv')
```

```python
data = [data2023_08, data2023_09, data2023_10]
df = pd.concat(data)
print(df.shape)
df.describe()
df
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
tmp=df[df['ownership'] == 'udziaÅ‚']
print(tmp)
df=df.drop(index=tmp.index).reset_index(drop='true')
df
```

```python
columns = df.columns
complete_data=[]
for i in columns:
  tmp = len(df[df[i].isnull()])/len(df)*100
  complete_data.append(tmp)

import matplotlib.pyplot as plt
plt.figure(figsize = (10, 10))
plt.barh(columns,complete_data )
plt.title('Unompleteness of data')
plt.xlabel('Features')
plt.ylabel('Percentages of null data')

```

```python
#df.drop(['id', 'condition', 'buildingMaterial'], axis=1, inplace=True)
df=df.drop([ 'condition', 'buildingMaterial'],axis=1).reset_index(drop='true')
df = df.dropna()


df
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = df.select_dtypes(['object']).columns
cat_cols=cat_cols.drop(['type','city','ownership','id'])

df_cats = df[cat_cols].reset_index(drop=True)
cat_cols
df[df_cats.columns] = df[df_cats.columns].apply(LabelEncoder().fit_transform)
df=df.drop([ 'id'],axis=1).reset_index(drop='true')
df
```

```python
df['priceForSquareMeters']=df['price']/df['squareMeters']
df
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

encoder = OneHotEncoder(sparse_output=False)
encoder1 = encoder.fit_transform(df[['city']])
df = pd.concat([df, pd.DataFrame(encoder1, columns=encoder.categories_[0])], axis=1)
encoder = OneHotEncoder(sparse=False)
encoder1 = encoder.fit_transform(df[['type']])
df = pd.concat([df, pd.DataFrame(encoder1, columns=encoder.categories_[0])], axis=1)
encoder = OneHotEncoder(sparse=False)
encoder1 = encoder.fit_transform(df[['ownership']])
df = pd.concat([df, pd.DataFrame(encoder1, columns=encoder.categories_[0])], axis=1)
df
df.columns
```

```python
# X_train=X_train.drop(['squareMeters','price'],axis=1)
df=df.drop(['city','type','ownership'],axis=1)
df
#think about drop latitude longitude columns due to city localization
```

```python
from sklearn.model_selection import train_test_split
tmp1=[df['hasBalcony'],df['hasParkingSpace'],df['hasElevator']]
target=pd.DataFrame(tmp1).transpose()
data=df.drop(['hasElevator','hasParkingSpace','hasBalcony'],axis=1)
X_train, X_test,Y_train, Y_test = train_test_split(data,target,test_size=0.4,random_state=21)
X_valid, X_test, Y_valid, Y_test = train_test_split(X_test,Y_test,test_size=0.5,random_state=21)
print(len(X_train))
print(len(X_test))
print(len(X_valid))
print(len(Y_train))
X_train.describe()
target
```

```python
dataTrain=pd.concat([X_train, Y_train], axis=1, join='inner')
tmp1=dataTrain[dataTrain['hasElevator'] == 0]
tmp2=tmp1[tmp1['floorCount'] > 4]
dataTrain=dataTrain.drop(index=tmp2.index).reset_index(drop='true')
print(dataTrain)
Y_train=[dataTrain['hasBalcony'],dataTrain['hasParkingSpace'],dataTrain['hasElevator']]
Y_train=pd.DataFrame(Y_train).transpose()
print(Y_train)
```

```python
X_train=dataTrain.drop(['hasBalcony','hasParkingSpace','hasElevator'],axis=1)
print(X_train)
```

```python
from sklearn.preprocessing import (
MinMaxScaler,
StandardScaler,
)
standard_scaler = StandardScaler().set_output(transform="pandas")
#data_standardized = standard_scaler.fit_transform(data2)
minmax_scaler = MinMaxScaler().set_output(transform="pandas")
data_normalized_train = minmax_scaler.fit_transform(X_train)
data_normalized_valid = minmax_scaler.fit_transform(X_valid)
data_normalized_test = minmax_scaler.fit_transform(X_test)

```

```python
columns=X_train.columns
columns=columns[0:20]
columns=columns.drop(['hasSecurity','hasStorageRoom'])
print(columns)
data_standardized_train = standard_scaler.fit_transform(X_train[columns])
data_standardized_valid = standard_scaler.fit_transform(X_valid[columns])
data_standardized_test = standard_scaler.fit_transform(X_test[columns])
print(data_standardized_train)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
clf = MLPClassifier(hidden_layer_sizes=(40,10,10,10,10), random_state=21, max_iter=1000, solver='adam',activation='identity',learning_rate='constant',early_stopping=True)
clf.fit(X_train, Y_train)
y_pred_train = clf.predict(X_train)
y_pred = clf.predict(X_valid)
print(y_pred_train[:,0])
acc_train_elevator = accuracy_score(Y_train['hasElevator'], y_pred_train[:,2])
acc_val_elevator = accuracy_score(Y_valid['hasElevator'], y_pred[:,2])
acc_train_parking = accuracy_score(Y_train['hasParkingSpace'], y_pred_train[:,1])
acc_val_parking = accuracy_score(Y_valid['hasParkingSpace'], y_pred[:,1])
acc_train_balcony = accuracy_score(Y_train['hasBalcony'], y_pred_train[:,0])
acc_val_balcony = accuracy_score(Y_valid['hasBalcony'], y_pred[:,0])
print(acc_train_elevator,acc_val_elevator,acc_train_parking,acc_val_parking,acc_train_balcony,acc_val_balcony)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
clf = KNeighborsClassifier(n_neighbors=10,weights='uniform',algorithm='brute')
clf.fit(data_normalized_train, Y_train)
y_pred_train = clf.predict(data_normalized_train)
y_pred = clf.predict(data_normalized_valid)
acc_train_elevator = accuracy_score(Y_train['hasElevator'], y_pred_train[:,2])
acc_val_elevator = accuracy_score(Y_valid['hasElevator'], y_pred[:,2])
acc_train_parking = accuracy_score(Y_train['hasParkingSpace'], y_pred_train[:,1])
acc_val_parking = accuracy_score(Y_valid['hasParkingSpace'], y_pred[:,1])
acc_train_balcony = accuracy_score(Y_train['hasBalcony'], y_pred_train[:,0])
acc_val_balcony = accuracy_score(Y_valid['hasBalcony'], y_pred[:,0])
print(acc_train_elevator,acc_val_elevator,acc_train_parking,acc_val_parking,acc_train_balcony,acc_val_balcony)
```
