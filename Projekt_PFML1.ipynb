```python
import pandas as pd
data2023_08 = pd.read_csv('drive/MyDrive/apartments_pl_2023_08.csv')
data2023_09 = pd.read_csv('drive/MyDrive/apartments_pl_2023_09.csv')
data2023_10 = pd.read_csv('drive/MyDrive/apartments_pl_2023_10.csv')
```

```python
data2023_10.describe()
```

```python
data = [data2023_08, data2023_09, data2023_10]
df = pd.concat(data)
print(df.shape)
df.describe()
df
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
columns = df.columns
complete_data=[]
for i in columns:
  tmp = len(df[df[i].isnull()])/len(df)*100
  complete_data.append(tmp)

import matplotlib.pyplot as plt
plt.figure(figsize = (10, 10))
plt.barh(columns,complete_data )
plt.title('Unompleteness of data')
plt.xlabel('Features')
plt.ylabel('Percentages of null data')

```

```python
df.drop(['id', 'condition', 'buildingMaterial'], axis=1, inplace=True)
df = df.dropna()
```

```python
from sklearn.model_selection import train_test_split
#X = data.drop('Exercise Intensity', axis = 1)
#y = data['Exercise Intensity']
X_train, X_test = train_test_split(df,test_size=0.4,random_state=21)
X_valid, X_test = train_test_split(X_test,test_size=0.5,random_state=21)
print(len(X_train))
print(len(X_test))
print(len(X_valid))
X_train.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = X_train.corr()
plt.figure(figsize = (15, 10))
sns.heatmap(corr, cmap = 'Greens', annot = True, fmt = '.2f')
```

```python
import seaborn as sns
sns.set_theme()
sns.boxplot(X_train,x='hasElevator',y='floorCount')
```

```python
tmp=X_train[X_train['hasElevator'] == 'no']
tmp=tmp[tmp['floorCount'] > 4]
X_train=X_train.drop(index=tmp.index).reset_index(drop='true')
X_train
```

```python
from sklearn.preprocessing import LabelEncoder
cat_cols = X_train.select_dtypes(['object']).columns
cat_cols=cat_cols.drop(['type','city','ownership'])

df_cats = X_train[cat_cols].reset_index(drop=True)
cat_cols
X_train[df_cats.columns] = X_train[df_cats.columns].apply(LabelEncoder().fit_transform)
X_train
```

```python
X_train['priceForSquareMeters']=X_train['price']/X_train['squareMeters']
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

encoder = OneHotEncoder(sparse=False)
encoder1 = encoder.fit_transform(X_train[['city']])
X_train = pd.concat([X_train, pd.DataFrame(encoder1, columns=encoder.categories_[0])], axis=1)
encoder = OneHotEncoder(sparse=False)
encoder1 = encoder.fit_transform(X_train[['type']])
X_train = pd.concat([X_train, pd.DataFrame(encoder1, columns=encoder.categories_[0])], axis=1)
encoder = OneHotEncoder(sparse=False)
encoder1 = encoder.fit_transform(X_train[['ownership']])
X_train = pd.concat([X_train, pd.DataFrame(encoder1, columns=encoder.categories_[0])], axis=1)
X_train
```
