```python
import pandas as pd
data2023_08 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/apartments_pl_2023_08.csv')
data2023_09 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/apartments_pl_2023_09.csv')
data2023_10 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/apartments_pl_2023_10.csv')
```

```python
data = [data2023_08, data2023_09, data2023_10]
df = pd.concat(data)
print(df.shape)
df.describe()
df
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python

df['priceForSquareMeter']=df['price']/df['squareMeters']
df
```

```python
filtered_df_all = df.drop(df[(df['hasElevator'] == 'yes') | (df['hasParkingSpace'] == 'yes') | (df['hasBalcony'] == 'yes')].index)
filtered_df_bps = df.drop(df[(df['hasElevator'] == 'no') | (df['hasParkingSpace'] == 'yes') | (df['hasBalcony'] == 'yes')].index)
filtered_df_eps = df.drop(df[(df['hasElevator'] == 'yes') | (df['hasParkingSpace'] == 'yes') | (df['hasBalcony'] == 'no')].index)
filtered_df_eb = df.drop(df[(df['hasElevator'] == 'yes') | (df['hasParkingSpace'] == 'no') | (df['hasBalcony'] == 'yes')].index)
```
